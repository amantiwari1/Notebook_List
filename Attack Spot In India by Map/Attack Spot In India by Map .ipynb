```python
#READ CSV IN ATTACK SPOT BY USING longitude AND latitude 

loca = pd.read_csv(r"D:\python\PROJECTS\dataset\indiadata.csv",encoding = "ISO-8859-1")
```

```python
#FUNCTION MAKE PLOT MAPS BY USING SHP

def plot_map(sf, figsize = (11,9)):
    '''
    Plot map with lim coordinates
    '''
    plt.figure(figsize = figsize)
    id=0
    for shape in sf.shapeRecords():
        x = [i[0] for i in shape.shape.points[:]]
        y = [i[1] for i in shape.shape.points[:]]
        plt.plot(x, y, 'k')
        
        
#SHOW MAP IN SHP FILE        
plot_map(sf)
```

```python
#MAKE SCATTER TO plot_map In Which Spot ATTACK BY longitude AND latitude

plot_map(sf)
plt.scatter(loca.longitude, loca.latitude, c='r', s=15)
plt.title('The Terrorist Attack Spot Between 1970-2015.',fontweight="bold",fontsize = 20)
plt.savefig('output.png')
plt.show()
```
