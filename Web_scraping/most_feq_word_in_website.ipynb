```
from urllib import request
from bs4 import BeautifulSoup
import re 
from nltk.corpus import stopwords
import nltk

#nltk.download("all")

url = "https://en.wikipedia.org/wiki/Big_data"
https = request.urlopen(url)

datahttps = BeautifulSoup(https, "html.parser")

para = datahttps.findAll("p")
data = ""
for i in para:
  data += i.text

clean_data = re.sub(r"\[[0-9]*\]"," ", data)
clean_data = re.sub(r"\s+", " ", clean_data)
clean_data = re.sub("[^a-zA-Z]", " ", clean_data)
clean_data = re.sub(r"\s+", " ", clean_data)

data_clean = []

for i in clean_data.split():
  if i.lower() not in stopwords.words("english"):
    data_clean.append(i.lower())

most_feq = nltk.FreqDist(data_clean)

most_feq.plot(20)
```
